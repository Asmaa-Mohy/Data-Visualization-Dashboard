## Learning plot

```python
import plotly.graph_objects as go

def Learning_plot(train_sizes, train_scores, test_scores):
  # Create means and standard deviations of training set scores
  train_mean = np.mean(-train_scores, axis=1)
  train_std = np.std(-train_scores, axis=1)

  # Create means and standard deviations of test set scores
  test_mean = np.mean(-test_scores, axis=1)
  test_std = np.std(-test_scores, axis=1)

  y_upper = train_mean + (20*train_std)
  y_lower = train_mean - (20*train_std)
  
  
  fig = go.Figure([
      go.Scatter(
          x=train_sizes, y=train_mean,
          name="Training score", line=dict(color='red',dash='dash')
      ),
      go.Scatter(
          x=train_sizes, y=test_mean,
          name="Cross-validation score", line=dict(color='blue')
      )
  ])
  fig.add_traces([go.Scatter(x = train_sizes, y = y_upper,
                             mode = 'lines', line_color = 'rgba(0,0,0,0)',
                           showlegend = False),
                go.Scatter(x = train_sizes, y = y_lower,
                           mode = 'lines', line_color = 'rgba(0,0,0,0)',
                           fill='tonexty', fillcolor = 'rgba(255, 0, 0, 0.2)',showlegend = False)])
  fig.update_layout(title_text='<b>Learning Curve</b>',xaxis_title_text='Training dataset')
  return fig
```

## Preprocessing of Data

```python
def Data_Preprocessing(df):
  # convert bathrooms and floors to intger
  df['bathrooms']=np.floor(df['bathrooms']).astype('int64')
  df['floors']=np.floor(df['floors']).astype('int64')
  #convert zipcode to category  #waterfront,view,condition,
  df['zipcode']= df['zipcode'].astype('category')
  df['age']= 2022-df['yr_built']
  df.drop([ 'lat','long', 'date','yr_built','yr_renovated'],axis=1,inplace=True)
  drop_index = df[(df['bathrooms']>11)| (df['sqft_lot']>1.2*(10**6)) |
                  (df['sqft_above']>8000)&(df['price']<(3*(10**6)))|
                  (df['sqft_basement']>3000)&(df['price']<(7*(10**6)))|
                  (df['sqft_living']>12000)&(df['price']<(4*(10**6)))|
                  (df['sqft_living']>10000)&(df['price']<(6*(10**6)))].index

  df = df.drop(drop_index)
  numer_col = df.drop(['waterfront','view','condition','zipcode'],axis=1)
  temp = numer_col.drop(['age'],axis=1)
  df[temp.columns] = df[temp.columns].apply(lambda m: np.log1p(m))
  # encoding of zipzone
  # from sklearn.preprocessing import LabelEncoder
  # label_encoder = LabelEncoder()
  # df['zipcode'] = df[['zipcode']].apply(label_encoder.fit_transform)
  df=pd.get_dummies(data=df,columns=['zipcode'])
  from sklearn.preprocessing import StandardScaler
  X_feature = df.drop('price', axis=1)
  scaler = StandardScaler()
  scaler.fit(X_feature)
  scaled_df = pd.DataFrame(scaler.transform(X_feature),columns= X_feature.columns ,index=X_feature.index)
  X_feature = scaled_df
  y_target = pd.DataFrame(df['price'].copy())
  y_target=y_target.set_index(scaled_df.index)
  return X_feature,y_target
```

## Use PCA to extract Principle component

```python
def PCA(X_feature):
  from sklearn.decomposition import PCA
  pca=PCA(n_components=70)
  pca.fit(X_feature)
  exp_var_pca = pca.explained_variance_ratio_
  cum_sum_eigenvalues = np.cumsum(exp_var_pca)
  PCA_df = pd.DataFrame(pca.transform(X_feature),index=X_feature.index)
  print('shape of original data', X_feature.shape)
  print('shape of original data', PCA_df.shape)
  print('explained variance for each component',pca.explained_variance_ratio_)
  print('cumliative explained variance for each component',pca.explained_variance_ratio_.cumsum())
  return PCA_df,exp_var_pca,cum_sum_eigenvalues
```

```python
def Split_Data(X_feature,y_target):

  # spliting data into train and test 
  from sklearn.model_selection import train_test_split
  X_all, X_test, y_all, y_test = train_test_split(X_feature,y_target,test_size=0.2,random_state=42)
  print('shape of feature test data',X_test.shape,', shape target',y_test.shape)
  print('shape of feature train data',X_all.shape,', shape target',X_all.shape)

  # split validation set
  X_train, X_val, y_train, y_val = train_test_split(X_all,y_all,test_size=0.2,random_state=42)
  print('shape of feature validation data',X_val.shape,', shape target',y_val.shape)
  print('shape of feature train data',X_train.shape,', shape target',y_train.shape)
  return X_all, X_test, y_all, y_test , X_train, X_val, y_train, y_val

```

## Build Models

```python
from xgboost import XGBRegressor
from sklearn.linear_model import  SGDRegressor, Ridge, ElasticNet, Lasso
from sklearn.model_selection import cross_val_score,cross_validate
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

def XGBR_Model(X_train, X_val, y_train, y_val,X_test,y_test):
  xmodel=XGBRegressor(learning_rate=0.1,n_estimators=200)
  xmodel.fit(X_train, y_train)
  predictions = xmodel.predict(X_val)
  xmodel_rmse = np.sqrt(mean_squared_error(y_val, predictions)) # mean_squared_error(y_target, predictions, squared=False)
  print("cost dy root mean squared error",xmodel_rmse) 
  # score of model use r-squared
  print("score of model",xmodel.score(X_val, y_val))
  preds_test= xmodel.predict(X_test)
  preds_test = np.exp(preds_test)-1
  y_test_real=np.exp(y_test)-1
  error = np.sqrt(mean_squared_error(y_test_real, preds_test))

  return xmodel,error

def SGD_Model(X_train, X_val, y_train, y_val,X_test,y_test):
  sgd = SGDRegressor(learning_rate='constant',eta0=0.0001)
  sgd.fit(X_train, y_train)
  predictions = sgd.predict(X_val)
  sgd_rmse = np.sqrt(mean_squared_error(y_val, predictions)) # mean_squared_error(y_target, predictions, squared=False)
  print("cost dy root mean squared error",sgd_rmse) 
  # score of model use r-squared
  print("score of model",sgd.score(X_val, y_val))
  preds_test= xmodel.predict(X_test)
  preds_test = np.exp(preds_test)-1
  y_test_real=np.exp(y_test)-1
  error = np.sqrt(mean_squared_error(y_test_real, preds_test))
  return sgd,error
```

```python
def LinearRegressionModel(X_train, X_val, y_train, y_val,X_test,y_test):
  lin_reg = LinearRegression()
  lin_reg.fit(X_train, y_train)
# mean_squared_error(y_val,lin_reg.predict(X_val))
  predictions = lin_reg.predict(X_val)
  lin_rmse = np.sqrt(mean_squared_error(y_val, predictions)) # mean_squared_error(y_target, predictions, squared=False)
  print("cost dy root mean squared error", lin_rmse) 
# score of model use r-squared
  print("score of model",lin_reg.score(X_val, y_val)) 

  preds_test= lin_reg.predict(X_test)
  preds_test = np.exp(preds_test)-1
  y_test_real=np.exp(y_test)-1
  error= np.sqrt(mean_squared_error(y_test_real, preds_test))
  return lin_reg,error
```

```python
from sklearn.preprocessing import PolynomialFeatures
def PolynomialRegression(X_train, X_val, y_train, y_val,X_test,y_test,X_all):
  # Polynomial regression
 
  poly=PolynomialFeatures(degree=2,include_bias=False)
  poly_features_X_train = poly.fit_transform(X_train)
  poly_features_X_all = poly.fit_transform(X_all)
  poly_features_X_val = poly.fit_transform(X_val)
# print(poly_features_X_train2.shape)
# print(X_train2.shape)
  poly_reg = LinearRegression(normalize=True)
  poly_reg.fit(poly_features_X_train, y_train)
  predictions = poly_reg.predict(poly_features_X_val)
  poly_reg_rmse = np.sqrt(mean_squared_error(y_val, predictions)) # mean_squared_error(y_target, predictions, squared=False)
  print("cost dy root mean squared error",poly_reg_rmse) 
# score of model use r-squared
  print("score of model",poly_reg.score(poly_features_X_val, y_val))

  poly_features_X_test = poly.fit_transform(X_test)
  preds_test= poly_reg.predict(poly_features_X_test)
  preds_test = np.exp(preds_test)-1
  y_test_real=np.exp(y_test)-1
  error = np.sqrt(mean_squared_error(y_test_real, preds_test))
  return poly,poly_reg,error,poly_features_X_all
```

```python
from sklearn.svm import SVR
def SVRModel(X_train, X_val, y_train, y_val,X_test,y_test):
  
  svr=SVR()
  svr.fit(X_train, y_train)
  predictions = svr.predict(X_val)
  svr_rmse = np.sqrt(mean_squared_error(y_val, predictions)) # mean_squared_error(y_target, predictions, squared=False)
  print("cost dy root mean squared error",svr_rmse) 
# score of model use r-squared
  print("score of model",svr.score(X_val, y_val))

  preds_test= svr.predict(X_test)
  preds_test = np.exp(preds_test)-1
  y_test_real=np.exp(y_test)-1
  error= np.sqrt(mean_squared_error(y_test_real, preds_test))
  return svr,error
```
